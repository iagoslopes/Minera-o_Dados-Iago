# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
from numpy import random

x=random.randint(1,20, size=(10))

filter_arr = []

filter_arr = x % 2 == 0

newarr = x[filter_arr]

print('Array:', x)
print('\nElementos pares:', newarr)
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
import numpy as np

x=random.randint(50, size=(2,10))

newx = np.add(x[0], x[1])
newx1 = np.subtract(x[0], x[1])
newx2 = np.multiply(x[0], x[1])
newx3 = np.divide(x[0], x[1])

print('Array 1:', x[0],'\nArray 2:', x[1])
print('\nSoma\n', newx, '\n')
print('Subtração\n', newx1, '\n')
print('Multiplição\n', newx2, '\n')
print('Divisão\n', newx3)
```

# 07 - Criação de DataFrame com Dados Sintéticos

Crie um DataFrame fictício de vendas contendo colunas como Produto, Quantidade (entre 1 e 20), e Preço (entre 10 e 100).

```python
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
produtos_repetidos = random.choice(produtos, size=15)

quantidades = random.randint(1, 21, size=15)

precos = random.randint(10, 101, size=15)

df_vendas = pd.DataFrame({'Produto': produtos_repetidos, 'Quantidade': quantidades,'Preço': precos})
df_vendas
```

# 08 - Filtragem de Dados

Filtre o DataFrame criado no exercício anterior para mostrar apenas as vendas onde a quantidade é maior que 10.

```python
df_filtrado = df_vendas[df_vendas['Quantidade'] > 10]

df_filtrado
```

# 09 - Operações em Colunas

Adicione uma coluna ao DataFrame com o valor total da venda (Quantidade * Preço).

```python
passes  = pd.Series(df_vendas['Quantidade'] * df_vendas['Preço'])
df_vendas = pd.concat([df_vendas, passes.rename("Valor total")], axis=1)
df_vendas
```

# 10 - Agrupamento e Agregação

Agrupe o DataFrame pelo campo Produto e calcule a soma das quantidades vendidas para cada produto.

```python
df_agrupado = df_vendas.groupby('Produto')['Quantidade'].sum()

df_agrupado
```

# 14 - Gráfico de Dispersão

Crie um gráfico de dispersão que mostre a relação entre altura e peso em um conjunto de dados fictício.

```python
np.random.seed(0)
alturas = np.random.normal(170, 10, 100)
pesos = np.random.normal(70, 15, 100)

plt.scatter(alturas, pesos, alpha=0.7, color='lightblue', edgecolor='black')

plt.title('Relação entre Altura e Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Criando um DataFrame fictício com várias variáveis
np.random.seed(0)  # Para reprodutibilidade
dados = {
    'A': np.random.rand(100),  # 100 valores aleatórios entre 0 e 1
    'B': np.random.rand(100) * 10,  # 100 valores aleatórios entre 0 e 10
    'C': np.random.normal(50, 15, 100),  # 100 valores com distribuição normal (média 50, desvio 15)
    'D': np.random.randint(1, 100, 100),  # 100 valores inteiros aleatórios entre 1 e 100
    'E': np.random.uniform(0, 1, 100)  # 100 valores uniformemente distribuídos entre 0 e 1
}

df = pd.DataFrame(dados)

# Calculando a matriz de correlação
correlacao = df.corr()

# Criando o heatmap usando Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")

# Adicionando título
plt.title('Heatmap de Correlação entre Variáveis')

# Exibindo o heatmap
plt.show()

```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Criando um DataFrame fictício com várias variáveis
np.random.seed(0)  # Para reprodutibilidade
dados = {
    'A': np.random.rand(100),  # 100 valores aleatórios entre 0 e 1
    'B': np.random.rand(100) * 10,  # 100 valores aleatórios entre 0 e 10
    'C': np.random.normal(50, 15, 100),  # 100 valores com distribuição normal (média 50, desvio 15)
    'D': np.random.randint(1, 100, 100),  # 100 valores inteiros aleatórios entre 1 e 100
    'E': np.random.uniform(0, 1, 100)  # 100 valores uniformemente distribuídos entre 0 e 1
}

df = pd.DataFrame(dados)

# Criando o pairplot usando Seaborn
sns.pairplot(df, diag_kind='kde')

# Adicionando título
plt.suptitle('Pairplot de Variáveis Fictícias', y=1.02)

# Exibindo o pairplot
plt.show()
```
